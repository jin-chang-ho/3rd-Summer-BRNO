```python
# Assume that you have 2 lists of the same length "dice_a" and "dice_b". 
# Elements of the lists dice_a and dice_b represent the score of players a and b respectively. - e.g. score from 5th game for player a and b - dice_a[4], dice_b[4]
# Write a code (that iterates over those lists simultaneously) that determines the winner. 
# The winner is the player who won the most games, if both players won the same number of games, it is a draw.
# Player wins a game if a player gets a bigger score than the other player.  
# TIP: zip function may be useful
dice_a = [6, 5, 2, 3, 3, 2, 5, 2, 6, 5, 1, 3, 4] # player A
dice_b = [4, 2, 2, 5, 4, 6, 6, 6, 4, 2, 5, 1, 2] # player B

# Here write the code
count_a = 0
count_b = 0
for a, b in zip(dice_a, dice_b):
  if a > b:
    count_a += 1
  if a < b:
    count_b += 1

if count_a > count_b:
  print("A is winner!!")
elif count_a < count_b:
  print("B is winner!!")
else:
  print("Draw!!")
```
